### Connect rows belonging to same Document


```python
aa_docs = df[filt]
aa_docs.head(2)
```

```python
aa_small_docs = aa_docs.head(30)
aa_small_docs.head(2)
```

```python
from langchain_community.chat_models import ChatOllama
from langchain.schema import HumanMessage, SystemMessage, AIMessage

llm = ChatOllama(model="llama2")
```

```python
def is_from_same_document(page1, page2):
    try:
        system_message = """
        Your job is to determine whether the following two documents are from the same document or not. Please return a boolean value of True or False. 
        One of the key things to look out for is the same names, locations, or dates referred to in both documents.
        Another clear indicator is the continuation of the text from one document to the other.

        Document 1:
        {document1}

        Document 2:
        {document2}

        ---
        IMPORTANT: Make sure to only return the boolean value and nothing else
        """
        human_message = f"Document 1: {page1}\n\nDocument 2: {page2}"

        page1_example = aa_small_docs["transcription_text"][0]
        page2_example = aa_small_docs["transcription_text"][1]

        page3_example = aa_small_docs["transcription_text"][2]
        page4_example = aa_small_docs["transcription_text"][3]

        page5_example = aa_small_docs["transcription_text"][0]
        page6_example = aa_small_docs["transcription_text"][4]

        messages = [
            SystemMessage(content=system_message),
            HumanMessage(
                content=f"Document 1: {page1_example}\n\nDocument 2: {page2_example}"
            ),
            AIMessage(content="True"),
            HumanMessage(
                content=f"Document 1: {page3_example}\n\nDocument 2: {page4_example}"
            ),
            AIMessage(content="False"),
            HumanMessage(
                content=f"Document 1: {page5_example}\n\nDocument 2: {page6_example}"
            ),
            AIMessage(content="False"),
            HumanMessage(content=human_message),
        ]

        chat_model_response = llm.invoke(messages)

        return chat_model_response.content == "True"

    except Exception as e:
        print(e)
        return False
```

```python
page1 = aa_small_docs["transcription_text"].iloc[0]
page2 = aa_small_docs["transcription_text"].iloc[4]
is_from_same_document(page1, page2)
```

```python
test_df = pd.read_csv(
    "https://raw.githubusercontent.com/KeithGalli/historical-docs-analysis/master/data/task2-apprenticeship-agreements-sample.csv"
)
test_df.head(2)
```

```python
test_df_copy = test_df.copy().head(5)
merge_id = 0

for index, row in test_df_copy.iterrows():
    print(f"Currently processing index {index} of {len(test_df_copy)}")
    print(f"Current merge_id: {merge_id}")
    if index + 1 < len(test_df):
        page1 = test_df.loc[index, "transcription_text"]
        page2 = test_df.loc[index + 1, "transcription_text"]

        if is_from_same_document(page1, page2):
            test_df.loc[index, "merge_id"] = merge_id
            test_df.loc[index + 1, "merge_id"] = merge_id
            continue

        test_df.loc[index, "merge_id"] = merge_id
        merge_id += 1
```
