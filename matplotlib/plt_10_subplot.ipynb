```python
from data import dev_x, dev_y, py_dev_y, js_dev_y
import matplotlib.pyplot as plt

dev_x = dev_x()
dev_y = dev_y()
py_dev_y = py_dev_y()
js_dev_y = js_dev_y()
```

```python
# figure vs axis (subplots)
# the figure is the container holding our plots and axes (subplots)
# the axis is the actual plot (the area where the data is visualized)
# plt.style.use("fivethirtyeight")
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)
# sharex=True means that both plots will share the same x-axis if same x-axis is used (same column)
# sharey=True means that both plots will share the same y-axis if same y-axis is used (not used here) (same row)
# default is 1,1 (1 row, 1 column)

ax1.plot(dev_x, dev_y, label="All Devs", color="k", linestyle="--", marker=".")
ax1.plot(dev_x, py_dev_y, label="Python", color="b", marker="x")
# instead of using plt.legend() we can use ax1.legend(), ax1.set_title(), ax1.set_xlabel(), ax1.set_ylabel(), just add set_ before the function name

ax1.set_title("Median Salary (USD) by Age")
ax1.set_xlabel("Ages")
ax1.set_ylabel("Median Salary (USD)")
ax1.legend()
ax1.grid(True)

# second plot
ax2.plot(dev_x, js_dev_y, label="JavaScript", color="y", marker="o")
ax2.set_title("Median Salary (USD) by Age")
ax2.set_xlabel("Ages")
ax2.set_ylabel("Median Salary (USD)")
ax2.legend()
ax2.grid(True)

plt.tight_layout()
```

```python
fig2, ((ax3, ax4), (ax5, ax6)) = plt.subplots(2, 2, sharex=True, sharey=True)

ax3.plot(dev_x, dev_y, label="All Devs", color="k", linestyle="--", marker=".")
ax4.plot(dev_x, py_dev_y, label="Python", color="b", marker="x")
ax5.plot(dev_x, js_dev_y, label="JavaScript", color="y", marker="o")
ax6.plot(dev_x, dev_y, label="All Devs", color="k", linestyle="--", marker=".")

fig2.suptitle("Median Salary (USD) by Age")
ax3.set_title("All Devs")
ax4.set_title("Python")
ax5.set_title("JavaScript")
ax6.set_title("All Devs")

plt.tight_layout()
fig2.legend(loc="upper left")
```
