```python
from data import dev_x, dev_y, py_dev_y, js_dev_y
import matplotlib.pyplot as plt

dev_x = dev_x()
dev_y = dev_y()
py_dev_y = py_dev_y()
js_dev_y = js_dev_y()
```

```python
print(plt.style.available)
```

```python
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")

plt.style.use("dark_background")  # to use a style for the plot

plt.plot(
    dev_x,
    dev_y,
    color="y",
    linestyle="solid",
    linewidth=4,
    marker=".",
    label="All Devs",
)

plt.plot(
    dev_x,
    py_dev_y,
    color="#f00f00",
    linestyle="dotted",
    marker="o",
    linewidth=3,
    label="Python",
)

plt.plot(
    dev_x,
    js_dev_y,
    color="#0000ff",
    linestyle="dashdot",
    marker="x",
    label="JavaScript",
)

plt.tight_layout()  # to avoid text overlap

plt.grid(True)  # to show grid in the plot

plt.legend()  # legend is showed when we have labels in the plot

# plt.savefig("plot.png")  # to save the plot as an image
```
