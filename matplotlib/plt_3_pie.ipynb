```python
import matplotlib.pyplot as plt
import random

theme = random.choice(plt.style.available)
print(f"Using theme {theme}")
```

```python
plt.style.use(theme)

slices = [60, 40, 30, 20]

plt.title("PieChart")

plt.pie(
    slices,
    labels=["Sixty", "Forty", "Thirty", "Twenty"],
    colors=["#008fd5", "#fc4f30", "#e5ae37", "#6d904f"],
    startangle=90,
    explode=(0.1, 0.1, 0, 0),
    autopct="%.1f%%",
    wedgeprops={"edgecolor": "black"},
)

plt.tight_layout()
plt.show()
```

```python
data = [
    ("JavaScript", 59219),
    ("HTML/CSS", 55466),
    ("SQL", 47544),
    ("Python", 36443),
    ("Java", 35917),
    ("Bash/Shell/PowerShell", 31991),
    ("C#", 27097),
    ("PHP", 23030),
    ("C++", 20524),
    ("TypeScript", 18523),
    ("C", 18017),
    ("Other(s):", 7920),
    ("Ruby", 7331),
    ("Go", 7201),
    ("Assembly", 5833),
]

languages, counts = zip(*data)
print(len(languages), len(counts))
```

```python
plt.style.use("fivethirtyeight")
plt.title("Most Popular Languages")

explode = [0] * len(languages)
explode[:-6:-1] = [0.1] * 5

plt.pie(
    counts,
    labels=languages,
    explode=explode,
    autopct="%1.1f%%",
    wedgeprops={"edgecolor": "black"},
)
plt.tight_layout()
```
