```python
import matplotlib.pyplot as plt
from data import dev_x, dev_y, py_dev_y, js_dev_y

dev_x = dev_x()
dev_y = dev_y()
py_dev_y = py_dev_y()
js_dev_y = js_dev_y()
```

```python
import numpy as np

x_indexes = np.arange(len(dev_x))
width = 0.25

plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Median Salary (USD) by Age")
plt.style.use("seaborn-v0_8-dark")

plt.xticks(ticks=x_indexes, labels=dev_x)

plt.bar(x_indexes - width, dev_y, color="#444444", label="All Devs", width=width)
plt.bar(x_indexes, py_dev_y, color="#008fd5", label="Python", width=width)
plt.bar(x_indexes + width, js_dev_y, color="#e5ae38", label="JavaScript", width=width)

plt.legend()
```

```python
import csv
from collections import Counter

lang_counter = Counter()

with open("data.csv", "r") as f:
    csv_reader = csv.reader(f)
    next(csv_reader)

    for line in csv_reader:
        lang_counter.update(line[1].split(";"))
```

```python
lang_counter = lang_counter.most_common(15)
print(lang_counter)
```

```python
languages, count = zip(*lang_counter)
print(len(languages), len(count))
```

```python
# use the barh method to create a horizontal bar chart

plt.ylabel("Programming Languages")
plt.xlabel("Number of People Who Use")
plt.title("Most Popular Languages")

plt.style.use("seaborn-v0_8-dark")
plt.barh(languages, count, color="#444444", label="All Devs")
plt.legend()
plt.tight_layout()
```
