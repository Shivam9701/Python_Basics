```python
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use("seaborn-v0_8-notebook")

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5] # satisfaction level

plt.scatter(
    x, y, s=100, c=colors, edgecolor="black", linewidth=1, alpha=0.75, cmap="Greens"
)
cbar = plt.colorbar()
cbar.set_label("Satisfaction") # label of the color bar
# s = size of the dots, c = color or differentiator, edgecolor = edge color of marker, linewidth = edge width of marker, alpha = transparency, cmap = color map for the color of the dots
plt.show()
```

```python
data = pd.read_csv("scatter.csv")
view_count = data["view_count"]
likes = data["likes"]
ratio = data["ratio"]

plt.style.use("seaborn-v0_8-dark-palette")
plt.title("Trending YouTube Videos")
plt.xlabel("View Count")
plt.ylabel("Total Likes")

plt.xscale("log")
plt.yscale("log")  # log scale for both x and y axis

plt.scatter(
    view_count,
    likes,
    c=ratio,
    edgecolor="black",
    linewidth=1,
    alpha=0.75,
    cmap="summer",
)
cbar = plt.colorbar()
cbar.set_label("Like/Dislike Ratio")  # label of the color bar
```
