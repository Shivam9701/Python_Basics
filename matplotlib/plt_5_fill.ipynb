```python
from data import ages_x, py2_dev_y, js2_dev_y, dev2_y
import matplotlib.pyplot as plt
import pandas as pd


ages_x = ages_x()
py2_dev_y = py2_dev_y()
js2_dev_y = js2_dev_y()
dev2_y = dev2_y()

ages_x = pd.Series(ages_x)
py2_dev_y = pd.Series(py2_dev_y)
js2_dev_y = pd.Series(js2_dev_y)
```

```python
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages", fontdict={"fontsize": 15, "color": "red"})
plt.ylabel("Median Salary (USD)", fontdict={"fontsize": 15, "color": "red"})

plt.style.use("fivethirtyeight")

plt.plot(
    ages_x,
    py2_dev_y,
    color="y",
    linestyle="solid",
    linewidth=2,
    marker=".",
    label="All Devs",
)

plt.plot(
    ages_x,
    dev2_y,
    color="#f00f00",
    linestyle="dotted",
    marker="o",
    linewidth=1.8,
    label="Python",
)

# plt.plot([8], color="w", label="8", linestyle="dotted", marker="o")
overall_median = 57287

plt.fill_between(
    ages_x,
    py2_dev_y,
    overall_median,
    alpha=0.25,
    label="Python Salary above Median",
    where=(py2_dev_y > overall_median),
    interpolate=True,
)

plt.fill_between(
    ages_x,
    py2_dev_y,
    overall_median,
    alpha=0.25,
    color="g",
    label="Python Salary below Median",
    where=(py2_dev_y <= overall_median),
    interpolate=True,
)

# plt.fill_between( # fill between two lines , where the first line is less than the second line
#     ages_x,
#     py2_dev_y,
#     dev2_y,
#     alpha=0.25,
#     color="#f00f00",
#     label="Python Salary less",
#     where=(py2_dev_y <= dev2_y),
#     interpolate=True,
# )

plt.legend()  # legend is showed when we have labels in the plot
```
